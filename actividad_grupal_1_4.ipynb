```python
df_limpieza = df.copy() #copiar el dataframe a una version trabajable para preservar el original
```

```python
df_limpieza = df_limpieza.drop_duplicates() #funcion la cual borra las filas duplicadas
df_limpieza
```

```python
df_limpieza['age'] = df_limpieza['age'].fillna(df_limpieza['age'].mean()) #funcion la cual reemplaza las edades faltantes por la media de las edades
df_limpieza
```

```python
# Calcular el porcentaje de nulos
porcentaje_nulos = (df_limpieza.isnull().sum() / len(df_limpieza)) * 100

# Filtrar solo las que tienen nulos y mostrar ordenadas de mayor a menor
reporte_nulos = porcentaje_nulos[porcentaje_nulos > 0].sort_values(ascending=False)

print("Porcentaje de nulos por columna:")
print(reporte_nulos.map("{:.2f}%".format))
```

```python
#Debido a demasiados nulos en deck, eliminamos la columna
df_limpieza.drop(columns=['deck'], inplace=True)
```

```python
media_fare = df_limpieza['fare'].mean() #Media de los pasajes

desv_estandar_fare= df_limpieza['fare'].std() #Desviacion estandar de los pasajes

cut_off = desv_estandar_fare * 3 #medida de corte

lower, upper = media_fare - cut_off, desv_estandar_fare + cut_off #definir corte superior e inferior, 3 STD hacia arriba 3 STD hacia abajo

df_limpieza2 = df_limpieza[(df_limpieza['fare'] > lower) & (df_limpieza['fare'] < upper)]

df_limpieza2
```

```python
df_limpieza2.loc[:,'embarked'] = df_limpieza2['embarked'].fillna(df_limpieza2['embarked'].mode()[0])
df_limpieza2.loc[:,'embark_town'] = df_limpieza2['embark_town'].fillna(df_limpieza2['embark_town'].mode()[0])
```

```python
nulos_por_columna = df_limpieza2.isnull().sum()

print("Columnas con datos nulos y cantidad de filas afectadas:")
print(nulos_por_columna[nulos_por_columna > 0])
#Para comprobar que ya no quedan datos nulos
```

```python
preprocesamiento = ColumnTransformer(transformers=[
    ('num', StandardScaler(), ['age', 'parch','fare']), #Normalización
    ('cat', OneHotEncoder(drop="first",sparse_output=False), ['survived', 'sex','sibsp','embarked','class','who','adult_male','embark_town','alive','alone']) #Encoding
])

X_transf = preprocesamiento.fit_transform(df_limpieza2)
print(X_transf[0]) #Ejemplo de print de la primera fila.
```
